```python
from pyspark.sql import functions as F
```

```python
# RAW Path for retail
retail_raw_path = "/Volumes/project4cat/project4db/p4raw/retail_dataset.csv"

# Bronze output location for retail
retail_bronze_path = "/Volumes/project4cat/project4db/p4bronze/retail_bronze"

# RAW Path for customer
customer_raw_path = "/Volumes/project4cat/project4db/p4raw/customer_dataset.json"

# Bronze Path for customer
customer_bronze_path = "/Volumes/project4cat/project4db/p4bronze/customer_bronze"
```

```python
# Retail dataframe
df_retail_raw = (
    spark.read
    .option("header", "true")
    .option("inferSchema", "true")
    .csv(retail_raw_path)
)
# display(df_retail_raw)

# Custoemr dataframe
df_customer_raw = spark.read.json(customer_raw_path)
# display(df_customer_raw)
```

```python
# retail dataframe
df_retail_bronze = (
    df_retail_raw
    .withColumn("ingestion_timestamp",F.current_timestamp())
    .withColumn("source_file", F.lit("retail_dataset.csv"))
)
# display(df_retail_bronze)

# customer dataframe
df_customer_bronze = (
    df_customer_raw
    .withColumn("ingestion_timestamp",F.current_timestamp())
    .withColumn("source_file", F.lit("customer_dataset.json"))
)
# display(df_customer_bronze)
```

```python
# write retail dataframe
df_retail_bronze.write.format("delta").mode("overwrite").save(retail_bronze_path)

# write customer dataframe
df_customer_bronze.write.format("delta").mode("overwrite").save(customer_bronze_path)
```
